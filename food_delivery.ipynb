```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Weather']=df['Weather'].fillna(df['Weather'].mode()[0])
```

```python
df['Traffic_Level'].fillna(df['Traffic_Level'].mode()[0],inplace=True)
```

```python
df['Time_of_Day'].value_counts()
```

```python
df['Time_of_Day'].fillna(df['Time_of_Day'].mode()[0],inplace=True)
```

```python
df['Courier_Experience_yrs'].value_counts()
```

```python
df['Courier_Experience_yrs'].fillna(df['Courier_Experience_yrs'].median(),inplace=True)
```

```python
df=df.drop(columns="Order_ID",axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=['Weather','Traffic_Level','Time_of_Day','Vehicle_Type']
for i in cols:
  df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
df['Vehicle_Type'].corr(df['Delivery_Time_min'])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df.drop(columns='Vehicle_Type',inplace=True)
```

```python
df.head()
```

```python
X=df.drop(columns='Delivery_Time_min',axis=1)
y=df['Delivery_Time_min']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomForestRegressor()
```

```python
param_dist = {
    'n_estimators': [100, 300, 500, 800, 1000],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
rf_random = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
rf_random.fit(X_train, y_train)
```

```python
print("Best Parameters:", rf_random.best_params_)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

pred = rf_random.best_estimator_.predict(X_test)

print("MAE:", mean_absolute_error(y_test, pred))
print("R2:", r2_score(y_test, pred))
```

```python
import joblib

best_model = rf_random.best_estimator_
joblib.dump(best_model, "delivery_time_model.pkl")
```
